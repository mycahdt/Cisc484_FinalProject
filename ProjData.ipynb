```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as clrs
import pandas as pd
```

```python
'''
Converts the data as it is seen in projectResults.txt to an array
'''
data = open('/Users/briellehina/Desktop/UD/CISC484/Project/projectResults.txt', 'r')
fullData=[]
while True:
    line1= data.readline()
    line2=data.readline()
    if not line1:
        break
    if not line2:
        break
    line= line1+line2
    line= line.replace("\n", "")
    endParam= line.index(']') +1
    paramsStr= line[0:endParam]
    line= line[endParam:]
    startFScore= line.index('[')
    FScoreStr=line[startFScore:]
    line= line[0: startFScore]
    line=line.strip()
    strArr= line.split(" ")
    accuracy= float(strArr[0])
    loss= float(strArr[1])
    precision= float(strArr[2])
    recall= float(strArr[3])
    paramsStr= paramsStr.replace('[', "")
    paramsStr= paramsStr.replace(']', "")
    paramsStr= paramsStr.replace("'", "")
    params= paramsStr.split(" ")
    FScoreStr= FScoreStr.replace('[', "")
    FScoreStr= FScoreStr.replace(']', "")
    FScoreStr= FScoreStr.strip()
    FScore= FScoreStr.split(" ")
    fScores=np.array([])
    for f in FScore:
        if f != '':
            fScores= np.append(fScores,float(f))
    fScores= fScores.tolist()
    modelNum= int(params[0])
    splitRatio= float(params[1])
    learningRate= float(params[2])
    activationFunction= params[3]
    paramsList= [modelNum, splitRatio, learningRate, activationFunction]
    fullList= [paramsList, accuracy, loss, precision,recall, fScores]
    fullData.append(fullList)
print(fullData)
print(len(fullData))
```
